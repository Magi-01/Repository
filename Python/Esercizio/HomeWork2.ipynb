```python
#1.)
class Heap():
    def __init__(self):
        pass

    def MAXheapify(self,arr, N, i):
        largest = i  # Initialize largest as parent
        l = 2 * i + 1  # left = 2*i + 1
        r = 2 * i + 2  # right = 2*i + 2

        # See if left child of parent exists and is greater than parent
        if l < N and arr[largest] < arr[l]:
            largest = l

        # See if right child of parent exists and is greater than parent
        if r < N and arr[largest] < arr[r]:
            largest = r

        # Change parent, if needed
        if largest != i:
            arr[i], arr[largest] = arr[largest], arr[i]  # swap

            # Heapify the parent.
            Heap.MAXheapify(self, arr, N, largest)


    def MINheapify(self,arr, N, i):
        smallest = i  # Initialize smallest as parent
        l = 2 * i + 1  # left = 2*i + 1
        r = 2 * i + 2  # right = 2*i + 2

        # See if left child of parent exists and is less than parent
        if l < N and arr[smallest] > arr[l]:
            smallest = l

        # See if right child of parent exists and is less than parent
        if r < N and arr[smallest] > arr[r]:
            smallest = r

        # Change parent, if needed
        if smallest != i:
            arr[i], arr[smallest] = arr[smallest], arr[i]  # swap

            # Heapify the parent.
            Heap.MINheapify(self, arr, N, smallest)


    def MAXheapSort(self,arr):
        N = len(arr)

        # Build max heap.
        for i in range(N // 2 - 1, -1, -1):
            Heap.MAXheapify(self, arr, N, i)

        # Extract elements
        for i in range(N - 1, 0, -1):
            arr[i], arr[0] = arr[0], arr[i]  # swap
            Heap.MAXheapify(self, arr, i, 0)


    def MINheapSort(self, arr):
        N = len(arr)

        # Build min heap.
        for i in range(N // 2 - 1, -1, -1):
            Heap.MINheapify(self, arr, N, i)

        # Extract elements
        for i in range(N - 1, 0, -1):
            arr[i], arr[0] = arr[0], arr[i]  # swap
            Heap.MINheapify(self, arr, i, 0)
            
arr = [19, 33, 4, 61, 5, 38, -36, 21, 0]
Contruct = Heap()
Contruct.MAXheapSort(arr)
print("MAXHeap sort (recursion): " + str(arr))
arr = [19, 33, 4, 61, 5, 38, -36, 21, 0]
Contruct.MINheapSort(arr)
print("MINHeap sort (recursion): " + str(arr))
```

```python
#2.) 
# Per il zero livello ho d^0 in quanto un elemento genitore.
# Per il primo livello ho d^1 in quanto d figli
# Per il secondo livello ho d^2 in quanto d figli
# ...
# Per il k-esimo livello ho d^k figli
# La somma di tutti i nodi di ogni livello dev'essere uguale ad n:
# 1+d+d^2+d^3+...d^k = n con k l'altezza
# Questo é la serie geometrica (d^(k+1)-1)/(d-1) = n
# Da cui si ricava d^(k+1) = n(d-1)+1 -> k+1 = log_d(n(d-1)+1)
# -> k = log_d(n(d-1)+1) - 1

#1.)
class Heap():
    def __init__(self, d):
        self.d = d

    def MAXheapify(self, arr, N, i):
        largest = i  # Initialize largest as parent
        for j in range(1, self.d + 1):
            child = self.d * i + j
            if child < N and arr[largest] < arr[child]:
                largest = child

        # Change parent, if needed
        if largest != i:
            arr[i], arr[largest] = arr[largest], arr[i]  # swap
            # Heapify the parent.
            Heap.MAXheapify(self, arr, N, largest)

    def MINheapify(self, arr, N, i):
        smallest = i  # Initialize smallest as parent
        for j in range(1, self.d + 1):
            child = self.d * i + j
            if child < N and arr[smallest] > arr[child]:
                smallest = child

        # Change parent, if needed
        if smallest != i:
            arr[i], arr[smallest] = arr[smallest], arr[i]  # swap
            # Heapify the parent.
            Heap.MINheapify(self, arr, N, smallest)

    def MAXheapSort(self, arr):
        N = len(arr)

        # Build max heap.
        for i in range(N // self.d - 1, -1, -1):
            Heap.MAXheapify(self, arr, N, i)

        # Extract elements
        for i in range(N - 1, 0, -1):
            arr[i], arr[0] = arr[0], arr[i]  # swap
            Heap.MAXheapify(self, arr, i, 0)

    def MINheapSort(self, arr):
        N = len(arr)

        # Build min heap.
        for i in range(N // self.d - 1, -1, -1):
            Heap.MINheapify(self, arr, N, i)

        # Extract elements
        for i in range(N - 1, 0, -1):
            arr[i], arr[0] = arr[0], arr[i]  # swap
            Heap.MINheapify(self, arr, i, 0)


arr = [19, 33, 4, 61, 5, 38, -36, 21, 0]
d = int(input("Insert size of children: "))
heap = Heap(d)
heap.MAXheapSort(arr)
print("MAXHeap sort (d-ary): " + str(arr))
heap.MINheapSort(arr)
print("MINHeap sort (d-ary): " + str(arr))
```
